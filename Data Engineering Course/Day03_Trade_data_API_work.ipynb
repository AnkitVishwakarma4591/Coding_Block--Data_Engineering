```python
import random
import time
def generate_trade():
    symbols = ["AAPL", "TSLA", "GOOG"]
    sides = ["BUY", "SELL"]

    trade = {
        "trade_id": random.randint(1000, 9999),
        "symbol": random.choice(symbols),
        "side": random.choice(sides),
        "quantity": str(random.randint(1, 100)),
        "price": str(round(random.uniform(100, 500), 2))
    }

    return trade
```

```python
while True:
  trade = generate_trade()
  print(trade)
  time.sleep(2)
```

```python
quantity = float(trade['quantity'])
price = float(trade['price'])
pni = quantity * price
print(f"Calculated PNI (Quantity * Price): {pni:.2f}")
```

**For Limited records**

```python
all_trades = []
num_trades = 10 # Define how many trades to generate
for _ in range(num_trades):
  trade = generate_trade()
  all_trades.append(trade)

print(f"Generated {len(all_trades)} trades.")
```

```python
total_pni = 0.0
for trade_item in all_trades:
    quantity = float(trade_item['quantity'])
    price = float(trade_item['price'])
    pni = quantity * price
    total_pni += pni

print(f"Total PNI for all {len(all_trades)} trades: {total_pni:.2f}")
```
