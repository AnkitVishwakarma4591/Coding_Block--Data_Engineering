Data Fetch Using Urls

```python
import requests

carts_url    = "https://fake-store-api.mock.beeceptor.com/api/carts"
orders_url   = "https://fake-store-api.mock.beeceptor.com/api/orders"
products_url = "https://fake-store-api.mock.beeceptor.com/api/products"

products = requests.get(products_url).json()
carts    = requests.get(carts_url).json()
orders   = requests.get(orders_url).json()
```

Task 1: Each Product Price after appling discount

```python
price_map = {}
for p in products:
    discount   = p.get('discount', 0)
    final_price = p['price'] - (p['price'] * discount / 100)
    price_map[p['product_id']] = final_price
    print(f"Product {p['product_id']}: ${p['price']} - {discount}% = ${final_price:.2f}")
```

Task 2 : For each Cart items total price (qty * discounted price)

```python
for cart in carts:
    cart_total = 0
    for item in cart['items']:
        unit_price  = price_map[item['product_id']]
        item_total  = unit_price * item['quantity']
        cart_total += item_total
        print(f"Product {item['product_id']} x Quantity {item['quantity']} @ ${unit_price:.2f} = ${item_total:.2f}")
    print(f"Cart {cart.get('cart_id')} TOTAL: ${cart_total:.2f}\n")
```

Task 3 : Cross-validate with Delivered orders

```python
for order in orders:
    if order['status'] == 'Delivered' and order['total_price'] > 0:
        calc_total = sum(
            price_map[i['product_id']] * i['quantity']
            for i in order['items']
        )
        diff  = abs(order['total_price'] - calc_total)
        match = diff < 1
        print(f"Order #{order['order_id']}: Total_Price = {order['total_price']}, "
              f"Calculated_Price = {calc_total:.2f}, Match={match}")
```
