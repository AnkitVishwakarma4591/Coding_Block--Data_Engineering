```python
import pandas as pd
import requests
import time
```

```python
URL = "https://randomuser.me/api/"
```

```python
dataset = []
while True:
    res = requests.get(URL)
    # if res.status_code == 200:
    json_data = res.json()
    # print(json_data['results']) # Printing email to show progress
    dataset.append(json_data['results'])
    
    time.sleep(0.1)
```

```python
flattened_data = [item[0] for item in dataset]
```

```python
df = pd.json_normalize(flattened_data)
```

```python
df
```

```python
mask = (df['gender'] == 'female') & (df['dob.age'] < 25)
target_customers_df = df[mask].head(10)
```

```python
emails_list = target_customers_df['email'].tolist()
```

```python
if len(target_customers_df) >= 10:
    records_needed = target_customers_df.index[-1] + 1
else:
    records_needed = len(df)
    print(f"Warning: Only found {len(target_customers_df)} matches in the entire dataset.")
```

```python
print(f"Total records processed to find 10 targets: {records_needed}")
print("\nTarget Emails:")
for i, email in enumerate(emails_list, 1):
    print(f"{i}. {email}")
```
