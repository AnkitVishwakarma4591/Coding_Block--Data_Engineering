```python
import requests
```

```python
products = requests.get("https://fake-store-api.mock.beeceptor.com/api/products").json()
orders = requests.get("https://fake-store-api.mock.beeceptor.com/api/orders").json()
```

```python
total_Product = []
for product in products:


  single_pro_details = {} # Dictionary to store details
  # update Dictionary with basic details of product like product_id, product_name, actual_rating
  single_pro_details.update({'Product_id' : product['product_id'], 'product_name' : product['name'], "Actual Rating" :product['rating']})


  # Average Rating for each Product
  total = 0
  cnt = 0
  ans = -1  # if their is no review on product
  if 'reviews' in product and product['reviews']:
    for rating in product['reviews']:
      total += rating['rating']
      cnt += 1
    ans = total/cnt
  else:
    ans = -1
  single_pro_details.update({'Average Rating' : ans})

  # Order Count
  product_id = product['product_id']
  order_count = 0
  for order in orders:
    for item in order['items']:
            if item['product_id'] == product_id:
                order_count += 1

  # update Dictionary with order_count
  single_pro_details.update({'Order Count' : order_count})

  # Product wise adding details in list
  total_Product.append(single_pro_details)
  print()
```

```python
for each_product in total_Product:
  print(each_product)
```
