# Data Structures for Data Engineering — Lists, Tuples, Sets, Dictionaries

This notebook covers:
- Lists
- Tuples
- Sets
- Dictionaries

All examples use data engineering style data such as customer IDs, orders, products, API responses, and event records.


## Day 8 — Lists


### 1. List basics


```python
customer_ids = [101, 102, 103, 104, 105]
print(customer_ids)
print(type(customer_ids))
```

### 2. Ordered, mutable, allows duplicates


```python
event_stream = ["login", "view_product", "add_to_cart", "view_product", "checkout"]
print(event_stream)

event_stream[1] = "search"
print(event_stream)
```

### 3. Indexing


```python
products = ["laptop", "phone", "watch", "tablet"]

print(products[0])
print(products[1])
print(products[-1])
print(products[-2])
```

### 4. Slicing


```python
orders = ["order_1001", "order_1002", "order_1003", "order_1004", "order_1005"]

print(orders[0:3])
print(orders[2:])
print(orders[:4])
print(orders[::2])
print(orders[::-1])
```

### 5. List of dictionaries


```python
orders = [
    {"order_id": 1001, "amount": 250.0, "status": "completed"},
    {"order_id": 1002, "amount": 499.0, "status": "pending"},
    {"order_id": 1003, "amount": 125.5, "status": "completed"}
]

print(orders)
print(orders[0])
print(orders[0]["amount"])
```

### 6. append()


```python
customer_ids = [101, 102, 103]
customer_ids.append(104)
print(customer_ids)
```

### 7. insert()


```python
customer_ids = [101, 103, 104]
customer_ids.insert(1, 102)
print(customer_ids)
```

### 8. extend()


```python
batch_1 = [1001, 1002, 1003]
batch_2 = [1004, 1005]

batch_1.extend(batch_2)
print(batch_1)
```

### 9. remove()


```python
statuses = ["completed", "pending", "failed", "pending"]
statuses.remove("pending")
print(statuses)
```

### 10. pop()


```python
api_batches = ["batch_1", "batch_2", "batch_3"]
last_batch = api_batches.pop()

print(last_batch)
print(api_batches)
```

### 11. sort()


```python
prices = [499.99, 199.50, 999.00, 299.99]
prices.sort()
print(prices)
```

### 12. reverse()


```python
daily_files = ["part_1.csv", "part_2.csv", "part_3.csv"]
daily_files.reverse()
print(daily_files)
```

### 13. Merging two lists of customer IDs


```python
customer_ids_source_a = [101, 102, 103, 104]
customer_ids_source_b = [103, 104, 105, 106]

merged_customer_ids = []
merged_customer_ids.extend(customer_ids_source_a)
merged_customer_ids.extend(customer_ids_source_b)

print(merged_customer_ids)
```

### 14. Removing duplicates manually from a list


```python
customer_ids = [101, 102, 103, 103, 104, 104, 105]
unique_customer_ids = []

for customer_id in customer_ids:
    if customer_id not in unique_customer_ids:
        unique_customer_ids.append(customer_id)

print(unique_customer_ids)
```

### 15. List memory view


```python
records = [1001, 1002, 1003]
print(records)
print(id(records))
print(id(records[0]))
print(id(records[1]))
```

## Day 9 — Tuples and Sets


### 16. Tuple basics


```python
schema = ("order_id", "customer_id", "amount", "status")
print(schema)
print(type(schema))
```

### 17. Tuple indexing


```python
partition_key = ("2026", "01", "sales")

print(partition_key[0])
print(partition_key[1])
print(partition_key[-1])
```

### 18. Tuple packing and unpacking


```python
record = ("AAPL", 100, 185.5)

symbol, quantity, price = record

print(symbol)
print(quantity)
print(price)
```

### 19. Tuple as immutable configuration


```python
db_connection = ("warehouse_host", 5432, "analytics_db")
print(db_connection)
```

### 20. Set basics


```python
customer_ids = {101, 102, 103, 104}
print(customer_ids)
print(type(customer_ids))
```

### 21. Convert list to set to remove duplicates


```python
messy_customer_ids = [101, 102, 102, 103, 104, 104, 105]
unique_customer_ids = set(messy_customer_ids)

print(messy_customer_ids)
print(unique_customer_ids)
```

### 22. Convert set back to list


```python
unique_customer_ids_list = list(unique_customer_ids)
print(unique_customer_ids_list)
```

### 23. Set add() and remove()


```python
active_users = {"u101", "u102", "u103"}
active_users.add("u104")
active_users.remove("u102")

print(active_users)
```

### 24. Union


```python
source_a = {101, 102, 103}
source_b = {103, 104, 105}

print(source_a | source_b)
print(source_a.union(source_b))
```

### 25. Intersection


```python
crm_users = {101, 102, 103, 104}
app_users = {103, 104, 105, 106}

print(crm_users & app_users)
print(crm_users.intersection(app_users))
```

### 26. Difference


```python
warehouse_users = {101, 102, 103, 104}
marketing_users = {103, 104}

print(warehouse_users - marketing_users)
print(warehouse_users.difference(marketing_users))
```

### 27. Symmetric difference


```python
stream_a = {"order_1", "order_2", "order_3"}
stream_b = {"order_3", "order_4", "order_5"}

print(stream_a ^ stream_b)
print(stream_a.symmetric_difference(stream_b))
```

### 28. Membership test in list vs set


```python
customer_list = list(range(1, 10001))
customer_set = set(customer_list)

target_id = 9999

print(target_id in customer_list)
print(target_id in customer_set)
```

### 29. Simple performance comparison: list vs set


```python
import time

customer_list = list(range(1, 200000))
customer_set = set(customer_list)
target_id = 199999

start = time.time()
result_list = target_id in customer_list
end = time.time()
print("List search:", result_list, "Time:", end - start)

start = time.time()
result_set = target_id in customer_set
end = time.time()
print("Set search:", result_set, "Time:", end - start)
```

## Day 10 — Dictionaries


### 30. Dictionary basics


```python
product = {
    "product_id": 1001,
    "name": "Laptop",
    "price": 799.99,
    "stock": 25
}

print(product)
print(type(product))
```

### 31. Accessing values by key


```python
product = {
    "product_id": 1001,
    "name": "Laptop",
    "price": 799.99,
    "stock": 25
}

print(product["name"])
print(product["price"])
```

### 32. Keys must be immutable


```python
product_catalog = {
    1001: "Laptop",
    1002: "Phone",
    1003: "Headphones"
}

print(product_catalog)
```

### 33. Updating dictionary values


```python
product = {
    "product_id": 1001,
    "name": "Laptop",
    "price": 799.99,
    "stock": 25
}

product["stock"] = 30
print(product)
```

### 34. Nested dictionaries


```python
api_response = {
    "order_id": 5001,
    "customer": {
        "customer_id": 101,
        "name": "Riya",
        "city": "Delhi"
    },
    "payment": {
        "method": "card",
        "status": "success"
    }
}

print(api_response)
print(api_response["customer"]["name"])
print(api_response["payment"]["status"])
```

### 35. Dictionary methods: keys(), values(), items()


```python
product = {
    "product_id": 1001,
    "name": "Laptop",
    "price": 799.99,
    "stock": 25
}

print(product.keys())
print(product.values())
print(product.items())
```

### 36. get()


```python
product = {
    "product_id": 1001,
    "name": "Laptop",
    "price": 799.99
}

print(product.get("name"))
print(product.get("stock"))
print(product.get("stock", 0))
```

### 37. update()


```python
product = {
    "product_id": 1001,
    "name": "Laptop",
    "price": 799.99
}

product.update({"stock": 25, "category": "electronics"})
print(product)
```

### 38. Looping through a dictionary


```python
product = {
    "product_id": 1001,
    "name": "Laptop",
    "price": 799.99,
    "stock": 25
}

for key, value in product.items():
    print(key, "->", value)
```

### 39. Product catalog dictionary


```python
product_catalog = {
    1001: {"name": "Laptop", "price": 799.99, "stock": 25},
    1002: {"name": "Phone", "price": 499.99, "stock": 50},
    1003: {"name": "Headphones", "price": 99.99, "stock": 100}
}

print(product_catalog)
print(product_catalog[1001])
print(product_catalog[1002]["price"])
```

### 40. Add a new product to catalog


```python
product_catalog = {
    1001: {"name": "Laptop", "price": 799.99, "stock": 25},
    1002: {"name": "Phone", "price": 499.99, "stock": 50}
}

product_catalog[1003] = {"name": "Keyboard", "price": 49.99, "stock": 80}
print(product_catalog)
```

### 41. Update stock in catalog


```python
product_catalog = {
    1001: {"name": "Laptop", "price": 799.99, "stock": 25},
    1002: {"name": "Phone", "price": 499.99, "stock": 50}
}

product_catalog[1001]["stock"] = 20
print(product_catalog)
```

### 42. Iterate over product catalog


```python
product_catalog = {
    1001: {"name": "Laptop", "price": 799.99, "stock": 25},
    1002: {"name": "Phone", "price": 499.99, "stock": 50},
    1003: {"name": "Headphones", "price": 99.99, "stock": 100}
}

for product_id, details in product_catalog.items():
    print(product_id, details["name"], details["price"], details["stock"])
```

### 43. API-style list of dictionaries and dictionary extraction


```python
products_api_response = [
    {"product_id": 1001, "name": "Laptop", "price": 799.99, "stock": 25},
    {"product_id": 1002, "name": "Phone", "price": 499.99, "stock": 50},
    {"product_id": 1003, "name": "Headphones", "price": 99.99, "stock": 100}
]

catalog = {}

for product in products_api_response:
    product_id = product["product_id"]
    catalog[product_id] = {
        "name": product["name"],
        "price": product["price"],
        "stock": product["stock"]
    }

print(catalog)
```

### 44. JSON-style event record


```python
event_record = {
    "event_id": "evt_1001",
    "event_type": "order_created",
    "payload": {
        "order_id": 5001,
        "customer_id": 101,
        "amount": 349.99
    },
    "metadata": {
        "source": "mobile_app",
        "timestamp": "2026-04-29T10:30:00Z"
    }
}

print(event_record["payload"]["order_id"])
print(event_record["metadata"]["source"])
```
