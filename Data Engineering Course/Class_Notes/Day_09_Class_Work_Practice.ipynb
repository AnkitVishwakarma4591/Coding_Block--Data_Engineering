Define a Product class with attributes as product_id, name, price etc - and suitable method to retrieve the details of the product

Call the products api to retrieve a list of products
Map the list of products to create objects for each product

```python
import requests
```

```python
products = requests.get('https://fake-store-api.mock.beeceptor.com/api/products').json()
```

```python
Product_id = []
Name = []
Price = []

for product in products:
    Product_id.append(product['product_id'])
    Name.append(product['name'])
    Price.append(product['price'])
```

```python
class Product_Details:
  def __init__(self, pro_id, pro_name, pro_price):
    self.pro_id = pro_id
    self.pro_name = pro_name
    self.pro_price = pro_price

  # def Product_Info(self):
  #   print(f"Product_Id : {self.pro_id}")
  #   print(f"Product_Name : {self.pro_name}")
  #   print(f"Product_Price : {self.pro_price}")

  def Product_Info(self):
    return (f"Product_Id : {self.pro_id} \n Product_Name : {self.pro_name} \n Product_Price : {self.pro_price}")
```

```python
for id, name, price in zip(Product_id, Name, Price):
   obj = Product_Details(id, name, price)
   print(obj.Product_Info())
   print("------")
```

Another way of doing

```python
products_objects = []
for id, name, price in zip(Product_id, Name, Price):
    obj = Product_Details(id, name, price)
    products_objects.append(obj)
```

```python
for product_obj in products_objects:
    product_obj.Product_Info()
    print("------")
```
