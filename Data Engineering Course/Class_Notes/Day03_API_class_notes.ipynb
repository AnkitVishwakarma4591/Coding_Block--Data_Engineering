```python
import requests
import pandas as pd
import time

URL = "https://randomuser.me/api/"
target_customers = []  # List to store the full dictionaries of matches
records_processed = 0

print("Searching for target customers...")

while len(target_customers) < 10:
    res = requests.get(URL)

    if res.status_code == 200:
        response_json = res.json()
        if 'results' in response_json and len(response_json['results']) > 0:
            records_processed += 1
            user_data = response_json['results'][0]

            # Condition 1: target_customer = Age < 25 and gender = female
            age = user_data['dob']['age']
            gender = user_data['gender']

            if age < 25 and gender == 'female':
                target_customers.append(user_data)
                print(f"Found {len(target_customers)}/10: {user_data['email']}")
        else:
            # Optional: handle cases where 'results' is empty or missing, e.g., log a warning
            pass

    time.sleep(0.1) # Avoid rate limiting

# Convert the list of results into a DataFrame
df = pd.json_normalize(target_customers)

print("\n--- Final Statistics ---")
print(f"Total records processed: {records_processed}")
print(f"Emails found: {len(df)}")

# Display the DataFrame
df[['email', 'gender', 'dob.age', 'nat']].head(10)
```
