## Polymorphism Example: Calculating Area of Different Shapes

```python
import math

class Shape:
    def area(self):
        pass
```

```python
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2
```

```python
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width
```

```python
# Demonstrating Polymorphism
shapes = [Circle(5), Rectangle(4, 6), Circle(10)]
```

```python
for shape in shapes:
    # The same method call works differently for each object
    print(f"The area of the {type(shape).__name__} is: {shape.area():.2f}")
```

## @abstractmethod Example

```python
from abc import ABC, abstractmethod

class Details(ABC):
    def __init__(self, name, branch, subject):
        self.name = name
        self.branch = branch
        self.subject = subject

    @abstractmethod
    def print_info(self):
        pass
```

```python
class Teacher(Details):
    def print_info(self):
        print("Teacher Details")
        print(f"Name: {self.name}\nBranch: {self.branch}\nSubject: {self.subject}")
```

```python
class Student(Details):
    def __init__(self, name, branch, subject, batch):
        super().__init__(name, branch, subject)
        self.batch = batch

    def print_info(self):
        print(f"Student Details")
        print(f"Name: {self.name}\nBranch: {self.branch}\nBatch: {self.batch}")
```

```python
t1 = Teacher("Mr. Umesh Sahu", "B.Tech", "Software Engineering")
t1.print_info()
```

```python
s1 = Student("Ankit", "B.Tech", "Software Engineering", "2023-27")
s1.print_info()
```
