```python
class Bank:
    def __init__(self, bank_name, balance=0):
        self.bank_name = bank_name
        self._balance = balance  # Changed to protected attribute by convention

    def get_balance(self):
        return self._balance

    def set_balance(self, value):
        if not isinstance(value, (int, float)) or value < 0:
            raise ValueError("Balance must be a non-negative number.")
        self._balance = value
```

```python
class User(Bank):
  def __init__(self,name,acc_no,bank_name="SBI"):
    super().__init__(bank_name)
    self.name = name
    self.acc_no = acc_no
    self.bank_name = bank_name

  def deposit(self,amount):
    current_balance = self.get_balance()
    self.set_balance(current_balance + amount)

  def withdraw(self,amount):
    current_balance = self.get_balance()
    self.set_balance(current_balance - amount)

  def check_balance(self):
    return self.get_balance()

  def user_info(self):
    print(f"Name: {self.name}")
    print(f"Bank : {self.bank_name}")
    print(f"Account Number: {self.acc_no}")
    print(f"Current Balance : {self.get_balance()}")
```

```python
user_01 = User("Ankit", 568996, "Kotak")
```

### Demonstrating User Class Functionality

```python
# Initial user info
print("Initial User Info:")
user_01.user_info()
```

```python
# Deposit some amount
deposit_amount = 1000
user_01.deposit(deposit_amount)
print(f"Deposited: {deposit_amount}")
print(f"Current Balance after deposit: {user_01.check_balance()}")
```

```python
# Withdraw some amount
withdraw_amount = 300
user_01.withdraw(withdraw_amount)
print(f"Withdrew: {withdraw_amount}")
print(f"Current Balance after withdrawal: {user_01.check_balance()}")
```

```python
# Final user info
print("Final User Info:")
user_01.user_info()
```
